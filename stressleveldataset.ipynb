```python
import pandas
data = pandas.read_csv("Datasets/StressLevelDataset.csv")
data
```

```python
# Describing the data
data.describe()
```

```python
# Separate the data
array = data.values
print(array)

X = array[:,0:20]
# print(X)
Y = array[:,20]
print(Y)
```

```python
# We split it further
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
# Import a module
from sklearn.svm import SVC 
model = SVC()
model.fit(X_train,Y_train)
```

```python
# Test the module
predictions =  model.predict(X_test)
print("Model Predicted:",predictions)
print("We expected:",Y_test)
```

```python
# Testing the accuracy
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(Y_test,predictions))
```

```python
suitability = [[12,22,0,6,3,1,4,4,3,2,1,3,6,2,4,3,5,4,2,2]]
outcome = model.predict(suitability)
print("Your stress level is:",outcome)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the countplot and assign it to ax
ax = sns.countplot(x="stress_level", data=data, palette="Reds")

# Titles and labels
plt.title("Distribution of stress levels")
plt.xlabel("Stress Level")
plt.ylabel("Number of Individuals")  

# Add values on top of bars
for container in ax.containers:
    ax.bar_label(container)

# Show plot
plt.show()
```

```python
# Below is a pieplot showing the percentage of people with different stress levels

data["stress_level"].value_counts().plot.pie(autopct='%1.1f%%')
```
