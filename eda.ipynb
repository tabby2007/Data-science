```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# use the described function
data.describe()
```

```python
# We have noticed that the column item price contains three empty slots
# we need to fill these empty slots with a value and in this case, we shall fill with the mean of the column
data["itemprice"] = data["itemprice"].fillna(55)

# describe the data again
data.describe()

```

```python
# Get the number of entries in a given column
data.groupby("gender").size()
```

```python
# combining multiple columns
data.groupby(["city","payment"]).size()
```

```python
# import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a count plot showing the distribution of male to female
sns.countplot(x="gender",data=data,palette="winter")
plt.title("Gender Distribution Against Total Count")
plt.ylabel("Total Count of Individuals")
plt.xlabel("Gender")
```

```python
# Class Work: Create a Count Plot of product
# Which product category appears the most?
sns.countplot(x="product", data=data, palette="Reds")
plt.xticks( rotation = 45)
plt.title("Distribution of products")
```

```python
# Below is a countplot that has two columns
sns.countplot(x="gender", hue="city", palette="Blues", data=data)
```

```python
# below is a pieplot
data["payment"].value_counts().plot.pie(autopct='%1.1f%%')
```
