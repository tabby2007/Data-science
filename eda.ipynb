```python
import pandas
data = pandas.read_csv("Datasets/StressLevelDataset.csv")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the countplot and assign it to ax
ax = sns.countplot(x="stress_level", data=data, palette="Reds")

# Titles and labels
plt.title("Distribution of stress levels")
plt.xlabel("Stress Level")
plt.ylabel("Number of Individuals")  

# Add values on top of bars
for container in ax.containers:
    ax.bar_label(container)

# Show plot
plt.show()
```

```python
# Below is a pieplot showing the percentage of people with different stress levels

data["stress_level"].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# A graph showing the distribution of sleep_quality
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(x="sleep_quality", data=data, palette="Reds")

plt.title("Distribution of sleep quality")
plt.xlabel("Sleep Quality")
plt.ylabel("Number of Individuals")

for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
# A graph showing the distribution of depression 
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(x="depression",data=data,palette="Blues")
plt.title("Distribution of depression levels")
plt.ylabel("Number of individuals")
plt.xlabel("depression")

for container in ax.containers:
    ax.bar_label(container,rotation = 90)
plt.show()
```

```python
# Below is a pieplot showing the distribution of people with mental health history
data["mental_health_history"].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# A line graph showing the distribution of the stress level against Peer pressure
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x="peer_pressure", y="stress_level", data=data, marker='o')

plt.title("Peer Pressure vs Stress Level")
plt.xlabel("Peer Pressure")
plt.ylabel("Stress Level")

plt.show()
```

```python
# A line graph showing the distribution of stress levels to social support
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x="social_support", y="stress_level", data=data, marker='o')

plt.title("Social Support vs Stress Level")
plt.xlabel("Social Support")
plt.ylabel("Stress Level")

plt.show()
```
